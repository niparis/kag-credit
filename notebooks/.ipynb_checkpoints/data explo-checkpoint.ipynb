## DebtRatio - percentage

Monthly debt payments, alimony,living costs divided by monthy gross income

This data points has an important threshold effect, as going above 1 means you're eating in your cash reserves

formula:
debt ratio = total expenses / monthly gross income

Note : means we can use this to compute expenses - that's the cash flow out, one critical element to answer our question (when do we run out of cash !)
Probably left out of the dataset on purpose (ie: seems like an obviously useful info)

```python
def perc80(x):
    return np.percentile(x.fillna(0), 80)
    
def perc20(x):
    return np.percentile(x.fillna(0), 20)

def mmean(x):
    return np.mean(x)
```

```python
# same check than for previous percentage.
# roughly same problem identified. Here the "typo" cutoff seems to be above 4
# it seems we need one regularization for the 4-1000 section, then a different rule for the 1000-10000
# the 3-4 bin has probably a mix of very high indebtness and typos. can they be separated ?
bins = [0, 0.5,0.8, 1, 1.25, 1.5, 2, 3, 4,5,6, 7, 8, 9, 10,100,1000,10000]
train['debt-ratio-bin'] = pd.cut(train['DebtRatio'], bins)
train.groupby('debt-ratio-bin').agg({'SeriousDlqin2yrs': 'mean', 
                                     'MonthlyIncome':('median', mmean, perc80, perc20), 
                                     'age': 'count'})
```

```python
# needed to keep this as a continuous var + better estimates of expenses
col = 'DebtRatio'
train.loc[(train[col] >=1000), col] = train.loc[(train[col] >=1000), col] /1000
train.loc[(train[col] >=100), col] = train.loc[(train[col] >=100), col] /100
train.loc[(train[col] >=4), col] = train.loc[(train[col] >=4), col] /10
```

```python
bins = [0, 0.5,0.8, 1, 1.25, 1.5, 2, 10,100,1000,10000]
train['debt-ratio-bin'] = pd.cut(train['DebtRatio'], bins)
train.groupby('debt-ratio-bin').agg({'SeriousDlqin2yrs': 'mean', 
                                     'age': 'count'})
```

```python
# income
train.loc[pd.isnull(train['MonthlyIncome']), 'MonthlyIncome'] = train['MonthlyIncome'].median()

# computing expenses
train['monthly-expenses'] = train['MonthlyIncome'] * train['DebtRatio']
# cash out
train['cash-out'] = train['MonthlyIncome'] - train['monthly-expenses']
```

```python
bins = [-np.inf, -10000, -100,0, 2500, 5000, 10000, np.inf]
train['cash-out-bin'] = pd.cut(train['cash-out'], bins)
train.groupby('cash-out-bin').agg({'SeriousDlqin2yrs': 'mean', 
                                     'age': 'count'})
```
